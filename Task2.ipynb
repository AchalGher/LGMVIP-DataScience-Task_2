# Before Splitting make copy of file

# Splitting the data set into train test data  sets

```python
target = Iris['Species']
Iris_1 = Iris.copy()
Iris_1 = Iris_1.drop('Species', axis =1)
```

# Defining the attributes

```python
x = Iris_1
```

```python
target
```

# Decision Tree Classifier

```python
y = target
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42 )
print("training split Input - ",x_train.shape)
print("training split Input - ",y_train.shape)
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train,y_train)
print("Decision Tree Classifier Created")
```

```python
y_pred = dtree.predict(x_test)
print('Classification Report - \n',classification_report(y_test,y_pred))
```

# ACCURACY RATE

```python
cm = confusion_matrix(y_test,y_pred)
plt.figure(figsize = (5,5))
sns.heatmap(data = cm ,linewidths = .5 , annot = True , square = True, cmap = 'Blues')
plt.ylabel('Actual label')
plt.xlabel('pedicted label')
all_sample_title = 'Accuracy Score : {0}'.format(dtree.score(x_test,y_test))
plt.title(all_sample_title,size = 15)
```

# TREE VISUALIZATION

```python
plt.figure(figsize =(20,20))
dec_tree = plot_tree(decision_tree = dtree , feature_names = Iris_1.columns,class_names = ["setosa", "vercicolor","verginica"],filled = True ,precision = 4 ,rounded = True)
```
